```python
from openai import OpenAI
import base64

client = OpenAI(api_key="你的API_KEY")

def encode_image(image_path):
    with open(image_path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

PROMPT = """Describe this artwork in detail. Include information about:
1. Visual content (figures, objects, spatial arrangement, light, color)
2. Historical context and authorship
3. Materials and technique
4. Cultural significance

Be specific and detailed."""

print("Generating descriptions...")

artwork_ids = ["caravaggio_calling", "raphael_athens", "rublev_trinity", "bhmpi_anonymous"]

for artwork_id in artwork_ids:
    image_path = f"data/images/{artwork_id}.jpg"
    print(f"Processing: {artwork_id}...")
    
    try:
        image_data = encode_image(image_path)
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[{
                "role": "user",
                "content": [
                    {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{image_data}"}},
                    {"type": "text", "text": PROMPT}
                ]
            }],
            max_tokens=800
        )
        description = response.choices[0].message.content
        
        output_path = f"data/descriptions/{artwork_id}.txt"
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(description)
        print(f"  Done: {len(description)} characters")
        
    except Exception as e:
        print(f"  Failed: {e}")

print("All done!")
```
